## Line equation

A frequently used approach to represent a straight line is to specify points $(x,y)$ on a line by:


$$
x = t + p \cdot y
$$

For given (constant) parameters of $t$ and $p$ values of $x,y$ are on a straight line. 

Rearranging this equation as:

$$
y = \frac{x}{p} - \frac{t}{p} 
$$

the slope $s$ of the line is 

$$
s = \frac{1}{p}
$$

and the offset $b$ is

$$
b = - \frac{t}{p}
$$

Considering a rectangular region shown below:

---



The region is bounded by points $\left( x_l, y_u \right)$, $\left( x_l, y_l \right)$, $\left( x_u, y_u \right)$ and $\left( x_u, y_l \right)$.

For specific parameter pairs $\left( t, p \right)$ the line intersects the rectangular region at points $\left(x_1, y_1 \right)$ and $\left(x_2, y_2 \right)$. Depending of parameters $\left( t, p \right)$ the following cases may occur:

1) no intersections

2) intersections

   a) $x_1 \le x_2$ and $y_1 \le y_2$

   b) $x_1 \le x_2$ and $y_2 \le y_1$

In the following section the intersection points are computed. Moreover the conditions for parameters $\left( t, p \right)$ to result in such intersections are studied.

From the inequality:

$$
x_l \le t + p \cdot y \le x_u
$$

the range of $y$ (assuming $p \neq 0$) is computed:

$$
\underbrace{\frac{x_l - t}{p}}_{y_{min}} \le y \le \underbrace{\frac{x_u - t}{p}}_{y_{max}}
$$

Intersections occur for:

$$
y_l \le y_{min} \le y_u
$$

or by inserting the expression for $y_{min}$:

 
$$
x_l - y_u \cdot p \le t \le x_l - y_l \cdot p 
$$

Hence for a given value of $p$ the range of $t$ for which intersections occur can be computed.
