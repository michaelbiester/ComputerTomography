## Projections

Assumptions:

image has $N_y$ in `y` direction and $N_x$ in `x` direction.

The center of a particular pixel defined by pixel indices $(n_x, n_y)$ shall be located at coordinates $n_y + 0.5$ and $n_x + 0.5$ . This is merely a convention. 

A line along which the projection shall be computed can be represented in different forms:



```python
N_x = 500
N_y = 800

N_fig_x = 700
N_fig_y = 1000
```

### Line equation

$$
y = m \cdot x + b
$$

The feasible range of `x`: $0 \le x \le N_x$

The feasible range of `y`: $0 \le y \le N_y$

$$
0 \le m \cdot x + b \le N_y
$$
$$
\left( \frac{-b}{m}\right) \le x \le \left( \frac{N_y - b}{m}\right)
$$

and taking into account the feasible range of `x` we obtain:

$$
x_{min} = max\left(0, \left(\frac{-b}{m}\right) \right) \le x \le x_{max} = min\left( N_x, \left( \frac{N_y - b}{m}\right) \right)
$$

```python
# line equation

b = 100
m = -3.5

# compute x-range where y values are within rectangle
xMin = max(0, -b/m)
xMax = min(N_x, (N_y - b)/m)

y_xMin = m * xMin + b
y_xMax = m * xMax + b

print(f"xMin: {xMin} ; xMax: {xMax}")
print(f"y_xMin: {y_xMin} ; y_xMax: {y_xMax}")

```

```python
fig1 = plt.figure(1, figsize=[12, 8])
ax_f1 = fig1.add_subplot(1, 1, 1)
ax = plt.gca()
ax_f1.add_patch(pat.Rectangle( (0,0), width=N_x, height=N_y, edgecolor=None, facecolor='lightgrey'))
# ax_f1.legend()
ax_f1.grid(True)
ax_f1.set_xlabel('x')
ax_f1.set_ylabel('y')
ax_f1.set_title('figure');
ax_f1.set_xlim(0, N_fig_x)
ax_f1.set_ylim(0, N_fig_y);
```
